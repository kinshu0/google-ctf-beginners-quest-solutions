```python
# maximum index of f1[1]
im_in_question = f7

max_index = np.argmax(im_in_question[1])
min_index = np.argmin(im_in_question[1])
# max_index = 0
bg = Image.new(mode='RGB', size=(640,480))

# pixel manipulation
for i in range(max_index, max_index + 640*480):
    r = int((im_in_question[2][i] + 0.4) * 255)
    g = int((im_in_question[3][i] + 0.4) * 255)
    b = int((im_in_question[4][i] + 0.4) * 255)
    bg.putpixel(((i - max_index) % 640, (i - max_index) // 640), (r,g,b))

bg
```

```python

current = f7

img_frames = []
frame_index = 0

pixels_per_frame = 640*480

period = 100

i = 0

while i < len(current) - pixels_per_frame:
    img = np.zeros((pixels_per_frame,3), dtype=np.uint8)

    start = i
    stop = start + pixels_per_frame

    img[:,0] = ((current[2][start: stop] + 0.4) * 255).astype(np.uint8)
    img[:,1] = ((current[3][start: stop] + 0.4) * 255).astype(np.uint8)
    img[:,2] = ((current[4][start: stop] + 0.4) * 255).astype(np.uint8)

    
    img = img.reshape((480,640,3))
    img_frames.append(img)
    frame_index += 1
    print(f'{frame_index} done')

    i += period


```

```python
# frames to video
import cv2

fourcc = cv2.VideoWriter_fourcc(*'mp4v')

out = cv2.VideoWriter('feed7.mp4', fourcc, 30, (640,480))

for frame in img_frames:
    out.write(frame)

out.release()
```
