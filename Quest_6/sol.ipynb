```python
with open('chall.txt', 'rb') as f:
    raw_contents = f.read()
```

```python
chinese_char_decode = raw_contents.decode()
```

```python
char_map = {}
for i in chinese_char_decode:
    if i in char_map:
        char_map[i] += 1
    else:
        char_map[i] = 1
```

```python
len(char_map)
```

```python
alphabet = sorted(char_map.keys())
alphabet_map = {alphabet[i]: i for i in range(len(alphabet))}
```

```python
arr_data = []
for i in chinese_char_decode:
    arr_data.append(alphabet_map[i])
```
